```python
import psycopg2
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
def connect_to_db():
    cur = psycopg2.connect(
        database = "DBMS",
        user = "postgres",
        password = os.environ.get('PASSWORD'),
        host = "localhost"
    )
    return cur.cursor()
```

```python
cur = connect_to_db()
```

```python
cur.execute("""select * from "User";""")
```

```python
data = cur.fetchall()
```
